# Реализация модели

## Имплементация (6 баллов)

```python
import numpy as np
import pandas as pd
import warnings
from sklearn.base import BaseEstimator, ClassifierMixin

class MySGDClassifier(BaseEstimator, ClassifierMixin):
    """
    Предполагается, что в выборке всегда 2 класса
    """
    
    def __init__(self, C=1., alpha=0.1, max_epoch=50, batch_size=5):
        """
        C - коэф. регуляризации
        alpha - скорость спуска
        max_epoch - максимальное количество эпох
        """  
        self._C = C
        self._alpha = alpha
        self._max_epoch = max_epoch
        self._batch_size = batch_size
        self._w = None
        self._costs = None
    
    def sigmoid(self, z):
        return 1.0 / (1.0 + np.exp(-z))
    
    def cost_function(self, X, y, w):
        return -np.average(np.log(self.sigmoid(y * np.dot(X, w.T))), axis=0) + \
                    (self._C / (2 * X.shape[1])) * np.sum(np.power(w, 2), axis=0)
    
    def gradient(self, X, y, w): 
        return -np.average((y[:, np.newaxis] * X) * self.sigmoid(-y * np.dot(X, w.T))[:, np.newaxis], axis=0) + \
                (self._C / X.shape[1]) * (w * [False, True, True])
    
    def get_batches(self, perm, batch_size):
        batches_num = perm.shape[0] / batch_size + (1 if perm.shape[0] % batch_size else 0)
        return np.array_split(perm, batches_num)     
    
    def fit(self, X, y=None):
        '''
        Обучение модели
        '''
        X = np.c_[np.ones(X.shape[0]), X]
        y[y == 0] = -1
        y[y == 1] = 1
    
        self._w = np.random.uniform(-1.0 / X.shape[1], 1.0 / X.shape[1], X.shape[1])
        w_next = np.copy(self._w)
        self._costs = []
        for epoch in xrange(self._max_epoch):
            perm = np.random.permutation(X.shape[0])
            for batch_n, batch in enumerate(self.get_batches(perm, self._batch_size), 1):
                w_next -= (self._alpha / (epoch + 1)) * self.gradient(X[batch], y[batch], self._w)
                self._w = np.copy(w_next)
            self._costs.append(self.cost_function(X, y, self._w))
        return self
        
    def predict(self, X):
        '''
        Возвращение метки класса
        '''
        X = np.c_[np.ones(X.shape[0]), X]
        y_hat = np.sign(np.dot(X, self._w.T))
        y_hat[y_hat == -1] = 0
        y_hat[y_hat == 1] = 1
        return y_hat
    
    def predict_proba(self, X):
        '''
        Возвращение вероятности каждого из классов
        '''
        X = np.c_[np.ones(X.shape[0]), X]
        y = np.zeros((X.shape[0], 2), dtype=np.float64)
        y[:,1] = self.sigmoid(np.dot(X, self._w.T))
        y[:,0] = 1.0 - y[:,1]
        return y
```

```python
np.random.seed(0)

C1 = np.array([[0., -0.8], [1.5, 0.8]])
C2 = np.array([[1., -0.7], [2., 0.7]])
gauss1 = np.dot(np.random.randn(200, 2) + np.array([5, 3]), C1)
gauss2 = np.dot(np.random.randn(200, 2) + np.array([1.5, 0]), C2)

X = np.vstack([gauss1, gauss2])
y = np.r_[np.ones(200), np.zeros(200)]

plt.figure()
plt.title('Initial data')
plt.scatter(X[:,0], X[:,1], c=y)        
plt.show()

# ==========

clf = MySGDClassifier()
clf.fit(X,y)

plt.figure()
plt.title(''.join(('MySGDClassifier (roc_auc_score: ', str(roc_auc_score(y, clf.predict_proba(X)[:,1])), ')')))  
plt.scatter(X[:,0], X[:,1], c=clf.predict(X))
x = np.linspace(1,4,100)
plt.plot(x, -(clf._w[0] + x * clf._w[1]) / clf._w[2])
plt.show()

# ==========

clf2 = LogisticRegression()
clf2.fit(X,y)

plt.figure()
plt.title(''.join(('LogisticRegression (roc_auc_score: ', str(roc_auc_score(y, clf2.predict_proba(X)[:,1])), ')'))) 
plt.scatter(X[:,0], X[:,1], c=clf2.predict(X))
plt.show()
```

```python
for alpha in (1e-3, 1e-2, 1e-1, 0.5, 1.0, 2.0, 5.0):
    clf = MySGDClassifier(C=0.01, alpha=alpha, max_epoch=50, batch_size=5).fit(X, y)
    plt.plot(xrange(1, len(clf._costs) + 1), clf._costs, label=str(alpha))
plt.xlabel(u'epoch_number')
plt.ylabel(u'cost_function')
plt.legend()
```
