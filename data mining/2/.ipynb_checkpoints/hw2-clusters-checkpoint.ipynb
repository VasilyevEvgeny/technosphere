Алгоритм кластеризации должен удовлетворять следующему интерфейсу. Конструктор принимает набор параметров, необходимых для работы алгоритма кластеризации. Метод `fit` подсчитывает параметры модели и возвращает `self`. Метод `predict` возвращает вектор с индексами кластеров для поданных в него объектов `x`

```python
from sklearn.base import BaseEstimator, ClusterMixin
from scipy.spatial import distance
import itertools

class Clustering(BaseEstimator, ClusterMixin):
        
    def __init__(self, k, linkage='single'):
        self.k = k
        self.linkage = linkage
        self.clusters = {}
        
        return None
        
    def fit(self, x, y=None):
        
        # Initial condition: every element is a cluster
        for i in xrange(x.shape[0]):
            self.clusters.update({'cluster_' + str(i):[i]})
        
        
        # Clusters evolution cycle
        while len(self.clusters.keys()) > self.k:
            min_dist = +inf
            
            # For every pair of clusters...
            for key1, key2 in itertools.product(self.clusters.keys(), self.clusters.keys()):
                if key1 != key2:
                    pair_distances = []
                    
                    # ...we calculate distances between all elements...
                    for idx1, idx2 in itertools.product(self.clusters.get(key1), self.clusters.get(key2)):
                        idx1, idx2 = int(idx1), int(idx2)                    
                        pair_distances.append(distance.euclidean(x[idx1], x[idx2]))
                    
                    # ...and choose right linkage 
                    if self.linkage == 'single':
                        dist = min(pair_distances)
                    elif self.linkage == 'complete':
                        dist = max(pair_distances)
                    elif self.linkage == 'average':
                        dist = sum(pair_distances) / len(pair_distances)                   
                    
                    if dist < min_dist:
                        min_dist = dist
                        key1_join, key2_join = key1, key2
        
            # Clusters updating
            self.clusters[key1_join].extend(self.clusters[key2_join])
            del self.clusters[key2_join]
        
        
        
        return self
    
    def predict(self, x):
        """
        Using computed model parameters predict cluster
        for all objects from x
        """
        # Making list with labels of elements
        target = []
        for i in xrange(x.shape[0]):
            for key, key_num in zip(self.clusters.keys(), xrange(len(self.clusters.keys()))):
                for element in self.clusters[key]:
                    if i == element:
                        target.append(key_num)
                        break
         
        
        return target
    
    def fit_predict(self, x, y=None):
        self.fit(x, y)
        return self.predict(x)
```

Выбрать и реализовать критерий качества, на основании которого будут подбираться параметры модели. Варианты критериев:
- Инерция (сумма квадратов расстояний от каждой из точек до ближайшего к этой точке центроида)
- Средний диаметр (максимальное расстояние между точками в кластере) или радиус (расстояние от центроида до самой дальней от него точки в кластере)
- Sihouette

**Критерий качества необходимо выбрать таким образом, чтобы он подходил для реализованного алгоритма**

```python
def quality(x, y):
    """
    Silhouette
    """
    clusters = {}
    for i in xrange(x.shape[0]):
        if str(y[i]) not in clusters.keys():
            clusters.update({str(y[i]):[i]})
        else:
            clusters[str(y[i])].append(i)
    
    silhouette = [] 
    for i in xrange(x.shape[0]):
        b = []
        for key in clusters.keys():   
            par = 0.0
            for j in clusters[key]:
                par += distance.euclidean(x[i], x[j]) 
            
            if key == str(y[i]):
                if len(clusters[key]) > 1:
                    a = par / (len(clusters[key]) - 1)
                else:
                    a = par / len(clusters[key])
            else:
                b.append(par / len(clusters[key]))

        silhouette.append((min(b) - a) / max(a,min(b)))   
    
    return np.mean(silhouette)
```

Применим критерий для выбора параметров алгоритма. Предлагается изменить следующий код таким образом, чтобы кластеризация вызывалась с верным набором параметров. На графике можно будет увидеть зависимость критерия от параметров алгоритма. Необходимо выбрать оптимальные значения этих параметров.

```python
from sklearn.metrics import silhouette_score

ks = range(2, 21)
criteria = np.zeros(len(ks))
criteria_theor = np.zeros(len(ks))

for i, k in enumerate(ks):
    cls = Clustering(k)
    y = cls.fit_predict(x)
    criteria[i] = quality(x, y)
    criteria_theor[i] = silhouette_score(x, y, metric="euclidean")
    
pl.figure(figsize=(8, 6))
pl.plot(ks, criteria)
pl.plot(ks, criteria_theor)
pl.title("Silhouette")
pl.ylabel("Criteria Silhouette")
pl.xlabel("Number of clusters $k$")
pl.grid()
pl.show()
```
